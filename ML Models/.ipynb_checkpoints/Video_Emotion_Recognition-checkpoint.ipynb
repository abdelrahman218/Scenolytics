# Video Emotion Recognition Model Training Pipeline

## Configuration

### Imports

```python
import os
import cv2
import numpy as np
from retinaface import RetinaFace #Face Detection Algorithm
import tensorflow as tf
from tensorflow import keras
from keras import layers, Model
from keras.applications import EfficientNetB0
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

### Global Variables

```python
VIDEOS_DIR = os.path.abspath('../../../Datasets/Video')
EMOTIONS = { "01": "neutral", "02": "calm", "03": "happy", "04": "sad", "05": "angry", "06": "fearful", "07": "disgust", "08": "surprised"}
NUM_EMOTIONS_CLASSES = len(EMOTIONS)
    
# Video preprocessing
FRAMES_PER_VIDEO = 16  # Number of frames to extract
IMG_SIZE = 224  # EfficientNet input size
FRAME_RATE = 10  # Target FPS (will sample from original)
    
# Model
CNN_FEATURE_DIM = 1280  # EfficientNetB0 output
LSTM_UNITS = 256
LSTM_LAYERS = 2
DROPOUT_RATE = 0.4
    
# Training
BATCH_SIZE = 8  # Adjust based on GPU memory
EPOCHS = 80
LEARNING_RATE = 1e-4
PATIENCE = 15
    
# Paths
MODEL_SAVE_PATH = 'models/best_video_emotion_model.h5'
WEIGHTS_PATH = 'models/model_weights.h5'
```

## Pre-Processing

### Loading Dataset

```python
def detect_and_crop_face(frame):
    # Convert frame BGR -> RGB for RetinaFace
    RGB_Frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Detect Faces
    Faces = RetinaFace.detect_faces(RGB_Frame)

    if len(Faces) > 0:
        First_Face = next(iter(Faces.values()))
        Facial_Area = First_Face["facial_area"]
        x1, y1, x2, y2 = Facial_Area

        # Add padding
        padding = 20
        x1 = max(0, x1 - padding)
        y1 = max(0, y1 - padding)
        x2 = min(frame.shape[1], x2 + padding)
        y2 = min(frame.shape[0], y2 + padding)

        # Crop face
        face = frame[y1:y2, x1:x2]
        return face

    # Return original frame if no face detected
    return frame
```

```python
def extract_frames_from_video(video_path, num_frames=16, target_size=(224, 224)):
    Cap = cv2.VideoCapture(video_path)
    Total_Frames = int(Cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    if Total_Frames < num_frames:
        # If video too short, we'll duplicate Frames
        Frame_Indices = np.linspace(0, Total_Frames - 1, num_frames, dtype=int)
    else:
        # Sample evenly across video
        Frame_Indices = np.linspace(0, Total_Frames - 1, num_frames, dtype=int)
    
    Frames = []
    for i in Frame_Indices:
        Cap.set(cv2.CAP_PROP_POS_FRAMES, i)
        IsReturned, Frame = Cap.read()
        Frame = np.array(Frame)
        
        if IsReturned:
            Frame = detect_and_crop_face(Frame)
            Frame = cv2.resize(Frame, target_size)
            Frame = cv2.cvtColor(Frame, cv2.COLOR_BGR2RGB)
            Frames.append(Frame)
        else:
            # If Frame read fails, duplicate last Frame
            if Frames:
                Frames.append(Frames[-1])
            else:
                Frames.append(np.zeros((target_size[0], target_size[1], 3)))
    
    Cap.release()
    return np.array(Frames)
```

```python
Video_Sequences = []
Emotion_Label = []
    
for Actor in os.listdir(VIDEOS_DIR):
    Actor_Path = os.path.join(VIDEOS_DIR, Actor)

    if not os.path.exists(Actor_Path):
        print(f"Warning: {Actor_Path} not found")
        continue
        
    Video_Files = [f for f in os.listdir(Actor_Path) 
                      if f.endswith('.mp4') & f.startswith('02')]
        
    print(f"Loading {len(Video_Files)} videos From File: {Actor}")

    for Video_File in Video_Files:
        Video_Path = os.path.join(Actor_Path, Video_File)
            
        try:
            frames = extract_frames_from_video(
                Video_Path, 
                num_frames=FRAMES_PER_VIDEO,
                target_size=(IMG_SIZE, IMG_SIZE)
            )
                
            Video_Sequences.append(frames)
            Emotion_Label.append(EMOTIONS.get(Video_Path[6:8]))
                
        except Exception as e:
            print(f"Error processing {Video_Path}: {e}")
            continue
    
Video_Sequences = np.array(Video_Sequences, dtype=np.float32) / 255.0  # Normalize
Emotion_Label = np.array(Emotion_Label)
    
print(f"\nDataset loaded: {len(Video_Sequences)} videos")
print(f"Shape: {Video_Sequences.shape}")
```

### Split Data Train, Validation, Testing

```python
Video_Sequences_Train, Video_Sequences_Temp, Emotion_Label_Train, Emotion_Label_Temp = train_test_split(
    Video_Sequences, Emotion_Label, test_size=0.25, random_state=42, stratify=Emotion_Label
)
Video_Sequences_Val, Video_Sequences_Test, Emotion_Label_Val, Emotion_Label_Test = train_test_split(
    Video_Sequences_Temp, Emotion_Label_Temp, test_size=0.5, random_state=42, stratify=Emotion_Label_Temp
)
    
print(f"\nDataset split:")
print(f"Train: {(len(Video_Sequences_Train))} videos")
print(f"Validation: {len(Video_Sequences_Val)} videos")
print(f"Test: {len(Video_Sequences_Test)} videos")
```

### Check Class Imbalance

```python
# Compute class weights
Class_Weights = compute_class_weight(class_weight='balanced', classes=EMOTIONS, y=Emotion_Label_Train)

# Plot
plt.figure()
Bars = plt.bar([str(c) for c in EMOTIONS], Class_Weights)

# Add text labels on top of bars
for Bar in Bars:
    Height = Bar.get_height()
    plt.text(Bar.get_x() + Bar.get_width()/2, Height, f"{Height:.2f}", ha='center', va='bottom')

plt.xlabel("Emotion")
plt.ylabel("Weight")
plt.tight_layout()
plt.show()
```

## Model Building & Training

### Building

Architecture: <br>
    1. EfficientNetB0 (pre-trained) extracts spatial features from each frame <br>
    2. Bi-LSTM processes temporal sequence <br>
    3. Attention mechanism focuses on important frames <br>
    4. Classification head <br>

```python
Model_Input_Layer = layers.Input(shape=(FRAMES_PER_VIDEO, IMG_SIZE, IMG_SIZE, 3))
# ========================================================================
# SPATIAL FEATURE EXTRACTION (CNN)
# ========================================================================
Base_CNN = EfficientNetB0(
        include_top=False,
        weights='imagenet',
        input_shape=(IMG_SIZE, IMG_SIZE, 3),
        pooling='avg'
    )
    
# Freeze early layers, fine-tune later layers
for Layer in Base_CNN.layers[:-30]:
    Layer.trainable = False
    
# TimeDistributed wrapper applies CNN to each frame
CNN_Features = layers.TimeDistributed(Base_CNN)(Model_Input_Layer)

# Model_Output_Layer: (batch, frames, 1280)
CNN_Features = layers.Dropout(DROPOUT_RATE)(CNN_Features)
    
# ========================================================================
# TEMPORAL MODELING (Bi-LSTM)
# ========================================================================
    
# First LSTM layer
LSTM_Out = layers.Bidirectional(
    layers.LSTM(LSTM_UNITS, return_sequences=True, dropout=DROPOUT_RATE)
)(CNN_Features)
    
# Second LSTM layer
LSTM_Out = layers.Bidirectional(
    layers.LSTM(LSTM_UNITS, return_sequences=True, dropout=DROPOUT_RATE)
)(LSTM_Out)
# output: (batch, frames, 2*lstm_units)
    
# ========================================================================
# ATTENTION_Layer MECHANISM
# ========================================================================
    
# Attention Layer: learn importance of each frame
Attention_Layer = layers.Dense(1, activation='tanh')(LSTM_Out)
Attention_Layer = layers.Flatten()(Attention_Layer)
Attention_Layer = layers.Activation('softmax')(Attention_Layer)
Attention_Layer = layers.RepeatVector(2 * LSTM_UNITS)(Attention_Layer)
Attention_Layer = layers.Permute([2, 1])(Attention_Layer)
# output: (batch, frames, 2*lstm_units) - Attention Layer weights
    
# Apply Attention Layer weights
attended = layers.Multiply()([LSTM_Out, Attention_Layer])
attended = layers.Lambda(lambda x: tf.reduce_sum(x, axis=1))(attended)
# output: (batch, 2*lstm_units) - weighted sum
    
# ========================================================================
# CLASSIFICATION HEAD
# ========================================================================
    
x = layers.Dense(512, activation='relu')(attended)
x = layers.BatchNormalization()(x)
x = layers.Dropout(DROPOUT_RATE)(x)
    
x = layers.Dense(256, activation='relu')(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(DROPOUT_RATE * 0.8)(x)
    
# Model Output layer
Model_Output_Layer = layers.Dense(NUM_EMOTIONS_CLASSES, activation='softmax')(x)
    
# Create model
Model = Model(inputs=Model_Input_Layer, outputs=Model_Output_Layer, name='Video_Emotion_Recognition')

# Compile model
Model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=LEARNING_RATE),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy', keras.metrics.TopKCategoricalAccuracy(k=2, name='top_2_accuracy')]
    )
    
print(Model.summary())
```

### Training

```python
callbacks = [
        ModelCheckpoint(
            MODEL_SAVE_PATH,
            monitor='val_accuracy',
            save_best_only=True,
            mode='max',
            verbose=1
        ),
        EarlyStopping(
            monitor='val_loss',
            patience=PATIENCE,
            restore_best_weights=True,
            verbose=1
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=5,
            min_lr=1e-7,
            verbose=1
        )
    ]

history = Model.fit(
        Video_Sequences_Train, Emotion_Label_Train,
        validation_data=(Video_Sequences_Val, Emotion_Label_Val),
        batch_size=BATCH_SIZE,
        epochs=EPOCHS,
        callbacks=callbacks,
        class_weight=Class_Weights,
        verbose=1
    )
```

## Model Evaluation

### Evaluation

```python
# Predictions
Model_Emotion_Prediction = Model.predict(Video_Sequences_Test, batch_size= BATCH_SIZE)
Model_Emotion_Prediction_Classes = np.argmax(Model_Emotion_Prediction, axis=1)
    
# Classification report
print("Classification Report:")
print(classification_report(Emotion_Label_Test, Model_Emotion_Prediction_Classes, target_names= EMOTIONS.values))
    
# Confusion matrix
cm = confusion_matrix(Model_Emotion_Prediction, Model_Emotion_Prediction_Classes)
    
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels= EMOTIONS.values,
                yticklabels= EMOTIONS.values)
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.show()
    
# Per-class accuracy
class_accuracy = cm.diagonal() / cm.sum(axis=1)
for i, emotion in enumerate(EMOTIONS):
    print(f"{emotion}: {class_accuracy[i]:.2%}")
```

### Visualization

```python

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
# Accuracy
axes[0].plot(history.history['accuracy'], label='Train Accuracy')
axes[0].plot(history.history['val_accuracy'], label='Val Accuracy')
axes[0].set_title('Model Accuracy')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(True)
    
# Loss
axes[1].plot(history.history['loss'], label='Train Loss')
axes[1].plot(history.history['val_loss'], label='Val Loss')
axes[1].set_title('Model Loss')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True)
    
plt.tight_layout()
plt.show()
```
